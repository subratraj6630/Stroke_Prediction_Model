```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
rfc=RandomForestClassifier()
rfr=RandomForestRegressor()
```

```python
df=pd.read_csv("stroke_risk_dataset_v2.csv")
```

```python
df
```

```python
df['gender'] = df['gender'].map({'Male': 0, 'Female': 1})
```

```python
df
```

```python
y=df[["stroke_risk_percentage","at_risk"]]
```

```python
x=df.drop(columns=["stroke_risk_percentage","at_risk"])
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.0002)
```

```python
y_train_c=y_train["at_risk"]
```

```python
y_train_r=y_train["stroke_risk_percentage"]
```

```python
y_test_c=y_test["at_risk"]
```

```python
y_test_r=y_test["stroke_risk_percentage"]
```

```python
model_c=rfc.fit(x_train,y_train_c)
```

```python
model_r=rfr.fit(x_train,y_train_r)
```

```python
from sklearn.metrics import accuracy_score
```

```python
# Training accuracy
train_accuracy = rfc.score(x_train, y_train_c) * 100

# Testing accuracy
test_accuracy = rfc.score(x_test, y_test_c) * 100

print(f"Training Accuracy: {train_accuracy:.2f}%")
print(f"Testing Accuracy: {test_accuracy:.2f}%")
```

```python
# Training accuracy
train_accuracy_r = rfr.score(x_train, y_train_r) * 100

# Testing accuracy
test_accuracy_r = rfr.score(x_test, y_test_r) * 100

print(f"Training Accuracy: {train_accuracy:.2f}%")
print(f"Testing Accuracy: {test_accuracy:.2f}%")
```
